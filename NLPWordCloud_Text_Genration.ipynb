```python
import pandas as pd
```

```python
df = pd.read_excel('JailerYTComments.xlsx', sheet_name = 1)
```

```python
comments = df['Comment']
comment_list = comments.tolist()
```

```python
with open('JailerComment.txt', 'w', encoding= 'utf-8') as file:
  for comment in comment_list:
    file.write(str(comment) + '\n')
```

```python
with open('JailerComment.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# tokenize the text
tokens = text.split()
tokens = [word.lower() for word in tokens]
```

```python
len(tokens)
```

```python
comment_word = {}
for token in tokens:
  if token not in comment_word.keys():
    comment_word[token] = 1
  else:
    comment_word[token] = comment_word[token] + 1
```

```python
comment_word = sorted(comment_word.items(), key=lambda item: item[1], reverse = True)
```

```python
comment_word = dict(comment_word)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a word cloud
wordcloud = WordCloud().generate_from_frequencies(comment_word)

# Display the generated image:
plt.imshow(wordcloud, interpolation='gaussian')
plt.axis("off")
plt.show()
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
token_no_stopwords = [word for word in tokens if not word in set(stopwords.words('english'))]  # Removing stopwords
```

```python
len(token_no_stopwords)
```

```python
comment_word_nostopwords = {}
for token in token_no_stopwords:
  if token not in comment_word_nostopwords.keys():
    comment_word_nostopwords[token] = 1
  else:
    comment_word_nostopwords[token] = comment_word_nostopwords[token] + 1
comment_word_nostopwords = sorted(comment_word_nostopwords.items(), key=lambda item: item[1], reverse = True)
comment_word_nostopwords = dict(comment_word_nostopwords)
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in comment_word_nostopwords]  # 'v' stands for verb
```

```python
len(lemmatized_words)
```

```python
lemmatized_words_nostopwords = {}
for token in token_no_stopwords:
  if token not in lemmatized_words_nostopwords.keys():
    lemmatized_words_nostopwords[token] = 1
  else:
    lemmatized_words_nostopwords[token] = lemmatized_words_nostopwords[token] + 1
lemmatized_words_nostopwords = sorted(lemmatized_words_nostopwords.items(), key=lambda item: item[1], reverse = True)
lemmatized_words_nostopwords = dict(lemmatized_words_nostopwords)
```
